```python
#定义打印格式
from langchain_core.messages import convert_to_messages

def pretty_print_message(message, indent=False):
    pretty_message = message.pretty_repr(html=True)
    if not indent:
        print(pretty_message)
        return

    indented = "\n".join("\t" + c for c in pretty_message.split("\n"))
    print(indented)

def pretty_print_messages(update, last_message=False):
    is_subgraph = False
    if isinstance(update, tuple):
        ns, update = update
        # skip parent graph updates in the printouts
        if len(ns) == 0:
            return

        graph_id = ns[-1].split(":")[0]
        print(f"Update from subgraph {graph_id}:")
        print("\n")
        is_subgraph = True

    for node_name, node_update in update.items():
        update_label = f"Update from node {node_name}:"
        if is_subgraph:
            update_label = "\t" + update_label

        print(update_label)
        print("\n")

        messages = convert_to_messages(node_update["messages"])
        if last_message:
            messages = messages[-1:]

        for m in messages:
            pretty_print_message(m, indent=is_subgraph)
        print("\n")
```

# 如何定义协调者的身份？单纯的LLM或者也是作为agent？
# 另外，协调者如何与agent进行沟通？如果通过路由函数但需要对提示工程进行严格设定路由词
# 协调者使用交接工具与其他agent进行沟通，因此协调者也是独立的agent

```python
from langgraph.types import Send
from langchain.tools import tool
from typing import Annotated
from langgraph.graph import StateGraph, MessagesState
from langgraph.prebuilt import InjectedState, ToolNode
from langgraph.types import Command

from langgraph.store.memory import InMemoryStore

def create_task_description_handoff_tool(
    *, agent_name: str, description: str | None = None
):
    name = f"transfer_to_{agent_name}"
    description = description or f"Ask {agent_name} for help."

    @tool(name, description=description)
    def handoff_tool(
        # this is populated by the supervisor LLM
        task_description: Annotated[
            str,
            "The task description of the next agent.",
        ],
        # these parameters are ignored by the LLM
        state: Annotated[MessagesState, InjectedState],
    ) -> Command:
        task_description_message = {"role": "user", "content": task_description}
        agent_input = {**state, "messages": [task_description_message]}
        return Command(
            goto=[Send(agent_name, agent_input)],
            graph=Command.PARENT,
        )

    return handoff_tool


assign_to_ML_agent_with_description = create_task_description_handoff_tool(
    agent_name="ML_agent",
    description="Assign oxidation prediction task to ML_agent.",
)

assign_to_Thermo_agent_with_description = create_task_description_handoff_tool(
    agent_name="Thermo_agent",
    description="Assign retrieve thermodynamic data task to Thermo_agent.",
)

assign_to_RAG_agent_with_description = create_task_description_handoff_tool(
    agent_name="RAG_agent",
    description="Assign retrieve scientific knowledge to RAG_agent.",
)

assign_to_Ref_agent_with_description = create_task_description_handoff_tool(
    agent_name="Ref_agent",
    description="Assign reference recommendation task to Ref_agent.",
)
```

```python
SUPERVISOR_SYSTEM_PROMPT = """
You are a **Supervisor Agent** managing four specialized assistants in the domain of **superalloy oxidation**.

🔒 **DOMAIN RESTRICTION**: Only respond to queries related to **superalloy and oxidation**. Ignore all other queries.

📌 **AGENT ROLES**:
1. **ML_agent** → Handle tasks involving prediction of oxidation mass/weight changes.
2. **Thermo_agent** → Handle tasks involving thermodynamic data of oxides (e.g., ΔG, pO₂, K(equilibrium constants).
3. **RAG_agent** → Handle tasks requiring background/general knowledge, explanations, or mechanism analysis.  
   ⚠️ Do not rewrite or alter RAG outputs — keep charts, figures, tables, **inline citation marks (e.g., [1], [2])**, and references exactly as returned.
   ⚠️ Avoid long single queries. You can break down long queries into subqueries and call the RAG_agent multiple times (labeled RAG1/2/3 results).
4. **Ref_agent** → Handle tasks recommending scientific references or literature.  
   ⚠️ Do not rewrite or alter Ref outputs — keep reference links and information exactly as returned.

📌 **THINKING STYLE**:
1. If the query is **simple**, directly assign it to the most appropriate agent.
2. If the query is **complex**, break it down step by step and assign subtasks to different agents in sequence, collecting their outputs.

📌 **SUPERVISOR ROLE**:
1. Always ensure that final answers from RAG_agent and Ref_agent retain **original formatting, links, charts, references, and citation markers**. 
2. When you are ready to provide the **final combined answer**, prefix it with: 'FINAL ANSWER'.
3. Your role is **coordination only**. Do not execute any tasks yourself.

===============================
EXAMPLES (Few-shot guidance)
===============================

Example 1 — Simple (ML only)
User: "Predict the oxidation weight gain of Co-9Al-7W alloy at 1000 °C for 10 h."
Supervisor plan:
- This is a quantitative prediction → ML_agent only.
Action:
→ Call ML_agent

---

Example 2 — Simple (RAG only)
User: "Why does Ti addition help continuous Al2O3 scale formation in Co-based alloys?"
Supervisor plan:
- Mechanism/background explanation → RAG_agent only.
Action:
→ Call RAG_agent with the exact question.
Note: Preserve inline citations, figures, and links. Do not rewrite or summarize.

---

Example 3 — Complex (ML + RAG + Thermo)
User: "Compare the oxidation resistance of Co-9Al-7W and Co-9Al-9W across 800–1100 °C for 100 h, and explain why."
Supervisor plan:
1. Use ML_agent to generate predictions at 800, 900, 1000, 1100 °C.
2. According to the ML results, Use RAG_agent to explain the mechanisms (such as: W role). 
3. If RAG mentions WO3 or spinel role, if need, call Thermo_agent to provide ΔG°, pO₂ at 800-1100°C.
Assembly Results:
FINAL ANSWER:
[ML prediction table]
[RAG explanation verbatim]
[Thermo data if applicable]
Supervisor conclusion: "···"
---

Example 4 — Complex (RAG + Ref)
User: "Find papers on reactive element effects in Ni-based superalloys"
Supervisor plan:
Call RAG_agent: "Summarize reactive element effects mechanisms"
Call Ref_agent: "Recommend papers on reactive elements in Ni-superalloys"
Assembly Results:
FINAL ANSWER:
[RAG explanation verbatim]
[Ref recommendations verbatim]
Supervisor conclusion: "···"

""".strip()
```

```python
from langgraph.prebuilt import create_react_agent
from langgraph_supervisor.handoff import create_forward_message_tool

forward_message = create_forward_message_tool("supervisor")

supervisor_agent_with_description = create_react_agent(
    model=llm_supervisor,
    tools=[
        assign_to_ML_agent_with_description,
        assign_to_Thermo_agent_with_description,
        assign_to_RAG_agent_with_description,
        assign_to_Ref_agent_with_description
    ],
    prompt=SUPERVISOR_SYSTEM_PROMPT,
    
    name="supervisor",
)
```

## 编译图

```python
from langgraph.graph import START, END
from langgraph.graph import StateGraph, START, MessagesState

supervisor = (
    StateGraph(MessagesState)
    .add_node(
        supervisor_agent_with_description, destinations=("ML_agent", "Thermo_agent", "RAG_agent", "Ref_agent",END)
    )
    .add_node(ML_agent)
    .add_node(Thermo_agent)
    .add_node(RAG_agent)
    .add_node(Ref_agent)
    
    .add_edge(START, "supervisor")
    # always return back to the supervisor
    .add_edge("ML_agent", "supervisor")
    .add_edge("Thermo_agent", "supervisor")
    .add_edge("RAG_agent", "supervisor")
    .add_edge("Ref_agent", "supervisor")
    .compile()
)
```

## 多智能体测试

```python
for chunk in supervisor.stream(
    {
        "messages": [
            {
                "role": "user",
                "content": "What is the oxidation weight gain of Co-9Al-9W at 900℃ for 100 hours? What roles do W and Al play in the oxidation process? Are they beneficial or harmful?"
            }
        ]
    },
    subgraphs=True,
):
    pretty_print_messages(chunk, last_message=False)
```

```python
for chunk in supervisor.stream(
    {
        "messages": [
            {
                "role": "user",
                "content": "Which one has better oxidation resistance in 1000 ℃, Co-9Al-9W or Co-9Al-9W-2Ti? Why? Can you explain the mechanism behind it?"
            }
        ]
    },
    subgraphs=True,
):
    pretty_print_messages(chunk, last_message=True)
```

# 在上述的多智能体测试中，我们询问一个相当抽象且复杂的问题：

问题：Co-9Al-9W 和 Co-9Al-9W-2Ti 两种合金在1000℃的氧化性能哪个更好，能否解释其背后的机理？

执行步骤：

1、协调者 首先调用机器学习智能体(ML_agent)自动默认预测 100h 的两种合金各自氧化增重做了对比，得出 Co-9Al-9W-2Ti 抗氧化性比 Co-9Al-9W 更好的初步基本理解；

2、协调者 随后调用检索增强生成智能体(RAG_agent)询问Ti的作用为何能促进合金的抗氧化性，以及针对 Co-9Al-9W-2Ti 和 Co-9Al-9W 具体合金做了提问；

3、协调者 将收集到所有的答案给出了综合性展示。
