```python
# Definition: Format and print the contents of multiple documents, with each document separated by --
#定义：格式化并打印多个文档的内容,每个文档用--隔开

def pretty_print(docs):
    print(
      f"\n{'-' * 100}\n".join([f"Document {i+1}:\n\n + {d.page_content}\n{d.metadata}" for i,d in enumerate(docs)])
  )
```

```python
import os
import json
from collections import defaultdict
from dotenv import load_dotenv

from langchain_community.chat_models import ChatZhipuAI
from langchain_community.embeddings import HuggingFaceBgeEmbeddings
from langchain_core.documents import Document
from langchain_community.vectorstores import FAISS

from transformers import AutoTokenizer

# =======================
# 初始化 LLM & Embedding & Tokenizer
# =======================
load_dotenv('C:/Users/12279/ZHIPU.env')
api_key = os.getenv('API_KEY')
llm = ChatZhipuAI(model="glm-4.5-flash", api_key=api_key)

model_path = r"D:\BGE_large_en_1.5v"
embeddings = HuggingFaceBgeEmbeddings(
    model_name=model_path,
    model_kwargs={'device': 'cuda', 'trust_remote_code': True},
    encode_kwargs={'normalize_embeddings': True}
)

tokenizer = AutoTokenizer.from_pretrained(model_path)

```

```python
input_folder = "D:/knowledge_base/Data_metadata/"
output_folder = "D:/knowledge_base/Data_ref_metadata/"
max_tokens = 435

all_documents = []  # ✅ 在外层统一存放所有 doc

for i in range(1, 101):  # 批量循环处理
    metadata_file = os.path.join(input_folder, f"metadata_{i}.json")

    if not os.path.exists(metadata_file):
        print(f"[WARN] 文件不存在，跳过：{metadata_file}")
        continue

    print(f"🚀 开始处理：{metadata_file}")

    with open(metadata_file, "r", encoding="utf-8") as f:
        items = json.load(f)

    grouped = defaultdict(dict)
    for item in items:
        src = item["metadata"]["source"]
        content_type = item["metadata"]["metadata_content"]
        content = item["page_content"]
        grouped[src][content_type] = content

    for source, parts in grouped.items():
        title = parts.get("document_title", {}).get("document_title", "")
        keywords_list = parts.get("keywords", {}).get("keywords", [])
        keywords = ", ".join(keywords_list)
        abstract = parts.get("abstract", {}).get("abstract", "")

        # ✅ 提取 DOI
        journal_info = parts.get("journal information", {}).get("journal information", [])
        doi = ""
        if isinstance(journal_info, list):
            for item in journal_info:
                if isinstance(item, dict) and "doi" in item:
                    doi = item["doi"]
                    break

        token_len = len(tokenizer.encode(abstract))
        if token_len > max_tokens:
            print(f"[INFO] Abstract too long ({token_len} tokens). Using ZhipuAI to compress...")
            prompt = f"Please compress the following English abstract to within 250 words, keeping the research aims, methods, and conclusions:\n\n{abstract}"
            shorter_abstract = llm.invoke(prompt).content.strip()
        else:
            shorter_abstract = abstract

        embedding_chunk = f"Title: {title}\nKeywords: {keywords}\nAbstract: {shorter_abstract}"

        # ✅ 每个 record 对应一个 doc
        doc = Document(
            page_content=embedding_chunk,
            metadata={
                "source": source,
                "title": title,
                "doi": doi
            }
        )
        all_documents.append(doc)

# ✅ 统一创建向量库
print(f"📦 一共收集到 {len(all_documents)} 篇文档")
vectordb = FAISS.from_documents(all_documents, embedding=embeddings)

# 保存索引
index_folder_path = output_folder
index_name = "faiss_metadata_all"
os.makedirs(index_folder_path, exist_ok=True)

try:
    print(f"开始保存合并索引：{index_name}")
    vectordb.save_local(folder_path=index_folder_path, index_name=index_name)
    print(f"✅ 索引已保存到 {index_folder_path}/{index_name}")
except Exception as e:
    print(f"❌ 保存失败：{e}")

print("🎉 所有文件已批处理完成！")
```

```python
import os, json
from collections import defaultdict

input_folder = "D:/knowledge_base/Data_metadata/"
output_folder = "D:/knowledge_base/Data_ref_metadata/"
max_tokens = 435

os.makedirs(output_folder, exist_ok=True)

all_documents = []          # ✅ 全量合并
combined_index = None       # ✅ 若想避免二次算向量，可逐批 add 到同一个索引

for i in range(1, 101):  # 批量循环处理
    metadata_file = os.path.join(input_folder, f"metadata_{i}.json")

    if not os.path.exists(metadata_file):
        print(f"[WARN] 文件不存在，跳过：{metadata_file}")
        continue

    print(f"🚀 开始处理：{metadata_file}")

    with open(metadata_file, "r", encoding="utf-8") as f:
        items = json.load(f)

    # --- 分组同一 source 的各段 ---
    grouped = defaultdict(dict)
    for item in items:
        src = item["metadata"]["source"]
        content_type = item["metadata"]["metadata_content"]
        content = item["page_content"]
        grouped[src][content_type] = content

    # --- 本批（单文件）文档列表，用于“分开保存”的索引 ---
    batch_documents = []

    for source, parts in grouped.items():
        title = parts.get("document_title", {}).get("document_title", "")
        keywords_list = parts.get("keywords", {}).get("keywords", [])
        keywords = ", ".join(keywords_list)
        abstract = parts.get("abstract", {}).get("abstract", "")

        # ✅ 提取 DOI
        journal_info = parts.get("journal information", {}).get("journal information", [])
        doi = ""
        if isinstance(journal_info, list):
            for jitem in journal_info:
                if isinstance(jitem, dict) and "doi" in jitem:
                    doi = jitem["doi"]
                    break

        # ✅ 长度与压缩（按你的原逻辑）
        token_len = len(tokenizer.encode(abstract))
        if token_len > max_tokens:
            print(f"[INFO] Abstract too long ({token_len} tokens). Using ZhipuAI to compress...")
            prompt = (
                "Please compress the following English abstract to within 250 words, "
                "keeping the research aims, methods, and conclusions:\n\n"
                f"{abstract}"
            )
            try:
                shorter_abstract = llm.invoke(prompt).content.strip()
            except Exception as e:
                print(f"[WARN] 压缩失败，使用原文摘要。原因：{e}")
                shorter_abstract = abstract
        else:
            shorter_abstract = abstract

        embedding_chunk = f"Title: {title}\nKeywords: {keywords}\nAbstract: {shorter_abstract}"

        doc = Document(
            page_content=embedding_chunk,
            metadata={
                "source": source,
                "title": title,
                "doi": doi
            }
        )
        batch_documents.append(doc)
        all_documents.append(doc)

    # === ① 保存“分开”的索引：每个 metadata_i.json 对应一个 ===
    if batch_documents:
        try:
            # 建这个批次的索引
            batch_db = FAISS.from_documents(batch_documents, embedding=embeddings)
            per_index_name = f"faiss_metadata_{i:03d}"  # 例如 faiss_metadata_001
            print(f"开始保存单文件索引：{per_index_name}")
            batch_db.save_local(folder_path=output_folder, index_name=per_index_name)
            print(f"✅ 单文件索引已保存到 {output_folder}/{per_index_name}")
        except Exception as e:
            print(f"❌ 保存单文件索引失败（metadata_{i}.json）：{e}")

        # === ② （可选优化）同时把本批文档 add 进合并索引，避免二次算向量 ===
        try:
            if combined_index is None:
                combined_index = FAISS.from_documents(batch_documents, embedding=embeddings)
            else:
                combined_index.add_documents(batch_documents)
        except Exception as e:
            print(f"[WARN] 合并索引增量添加失败（metadata_{i}.json）：{e}")

# === ③ 保存“合并”的索引 ===
print(f"📦 合并文档总计：{len(all_documents)} 篇")
try:
    if combined_index is None:
        # 兜底：如果上面没成功用增量构建，则一次性从 all_documents 构建
        combined_index = FAISS.from_documents(all_documents, embedding=embeddings)
    index_name = "faiss_metadata_all"
    print(f"开始保存合并索引：{index_name}")
    combined_index.save_local(folder_path=output_folder, index_name=index_name)
    print(f"✅ 合并索引已保存到 {output_folder}/{index_name}")
except Exception as e:
    print(f"❌ 保存合并索引失败：{e}")

print("🎉 所有文件已批处理完成！")
```

```python
# 保存 all_documents 为 JSON
save_json = os.path.join(output_folder, "all_documents.json")

# 把 Document 转成可序列化 dict
all_docs_serializable = [
    {
        "page_content": doc.page_content,
        "metadata": doc.metadata
    }
    for doc in all_documents
]

with open(save_json, "w", encoding="utf-8") as f:
    json.dump(all_docs_serializable, f, ensure_ascii=False, indent=2)

print(f"✅ all_documents 已保存到 {save_json}")
```
