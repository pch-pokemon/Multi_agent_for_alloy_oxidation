# 基本工具能力测试

导入工具 / Import the tool

```python
from ml import parse_alloy_composition,predict_mass_gain_one,predict_mass_gain_batch,predict_mass_gain_csv,evaluate_mass_gain_csv
```

(1) 解析合金式 / parse alloy composition

```python
parse_result = parse_alloy_composition.invoke("Co-10Al-5W-2Ti")
print(parse_result)
```

(2) 单次预测（可选工艺参数）/ Single prediction (optional heat treatment parameters)
包括/include: solu_temp, solu_temp, aging_temp1, aging_time1, aging_temp2, aging_time2

```python
prediction_result = predict_mass_gain_one.invoke("Co=85, Al=8, W=7, Temperature=800, Time=100, solu_temp=1000, solu_time=12")
print(prediction_result)
```

(3) 批量预测（每组预测按分号或换行隔开）/ Batch prediction (each prediction group is separated by a ';' or '\n')

```python
prediction_result = predict_mass_gain_batch.invoke(
    "Co=83.6, Al=8.1, W=8.3, Temperature=900, Time=10, solu_temp=1000, solu_time=12;\n"
    "Ni=70, Cr=20, Al=10, Temperature=1000, Time=50"
)
print(prediction_result)
```

（4） CSV 批量预测 & 评估 / CSV Batch Prediction & Evaluation

```python

from ml import p_data
sample = str(p_data("samples.csv"))
predict_mass_gain_csv.invoke(sample)
```

```python
evaluate_sample = str(p_data("samples_with_gt.csv"))
evaluate_mass_gain_csv.invoke(evaluate_sample)
```
